Напишите функцию принимающую на вход любое количество целочисленных параметров, и возвращающую их сумму 

```python
def summ(*args):
    res = 0
    for arg in args:
        if arg != int(arg):
            return print('Integer parameters required')
        else: res += arg
    return res
```

Построить график показывающий для каждой страны долю по доменам gmail.com и hotmail.com в процентах.

За исключением всех остальных доменов.

```python
data = pd.read_csv('df.csv', sep=';')
data.head()
```

```python
indeces_to_drop = []
for idx, email in enumerate(data['email']):
    if not (email.endswith('gmail.com') or email.endswith('hotmail.com')):
        indeces_to_drop.append(idx)
```

```python
data.drop(data.index[indeces_to_drop], inplace=True)
data['hue'] = data['email'].apply(lambda email: 1 if email.endswith('gmail.com') else 0)

subset = data.groupby(['country_code', 'hue']).size().unstack()
subset = (100. * subset.T / subset.T.sum()).T
```
