```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as sts
import random
%matplotlib inline

mpl.rcParams['savefig.dpi'] = 100
mpl.rcParams['figure.dpi'] = 100
```

#### Распределение хи-квадрат

```python
# герерируем 100 чисел равномерно распределенных от 0 до 20
x = np.linspace(0,20,100)

# герерируем хи-квадрат распределение
k = 3 # степени свободы хи-квадрат распределения
rv = sts.chi2(k)
pdf = rv.pdf(x)
# рисуем рлотность распределения
plt.plot(x, pdf, label="$k=%s$" % k)
plt.legend()
plt.title("PDF ($\chi^2_k$)")

# выбираем 1000 случайных величин из сгенерированного распределения
s = 1000
rvs = rv.rvs(size=s)
# строим гистограмму для этих величин
plt.hist(rvs, normed=True, bins=100)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Для хи-квадрат распределения матожидание равно числу k степеней свобод, а дисперсия - 2k.

Распределения для выборочного среднего с размером выборки n, согласно ЦПТ, должно стремиться к нормальному с матожиданием равному k и дисперсией равной 2k/n.

```python
# будем проводить расчеты и строить графики для 4х значений n - 5, 10, 50, 100, 500
nl = [5, 10, 50, 100, 500]
for n in nl:

    # создаем новый список с 1000 выборок
    o = []
    for smpl in range(1000):
        #rs = random.sample(rvs, n) # как вариант можно еще делать выборки из уже сделанной в предыдущем выборки rvs (не относится к упражнению)
        rs = rv.rvs(size=n)
        o.append(np.mean(rs))

    # строим гистограмму для полученных величин
    plt.hist(o, normed=True, bins=200)

    # герерируем 100 чисел равномерно распределенных от 0 до 10
    y = np.linspace(0,8,100)

    # геренируем нормальное распределение с параметрами mu = k и sigma = sqrt(2k/n)
    mu = k
    sigma = np.sqrt(2*float(k)/n)
    norm_rv = sts.norm(loc=mu, scale=sigma)
    pdf = norm_rv.pdf(y)
    plt.plot(y, pdf)

    print '### Выборка размером ', n, '###'
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend(('normal', 'sample'))
    plt.show()
    print 'Параметры исходного хи-квадрат распределения'
    print 'Mean:', rv.mean(), '\n', 'Var:', rv.var(), '\n', 'SD:', rv.std()
    print ''
    print 'Параметры распределения выборочных средних'
    print 'Mean:', np.mean(o), '\n', 'Var:', np.var(o), '\n', 'SD:', np.std(o)
    print ''
    print 'Параметры нормального распределения'
    print 'Mean:', norm_rv.mean(), '\n', 'Var:', norm_rv.var(), '\n', 'SD:', norm_rv.std()
    print ''
```

Как мы видим, чем больше значение выборки n, тем больше мы приближаемся к нормальному распределению. 
Величина матожидания равна матожиданию изначального распределения случайной величины, а дисперсия обратно пропорциональна значению выборки, по котрому мы производим усреднение.
