```python
#Я решил добавить pandas на всякий случай. Хорошая библиотека, но использовать я её, конечно, не буду.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
#Создаём выборку из 1000 элементов с распределением Рэлея и строим гистограмму по ней
distr = sts.rayleigh()
sample = distr.rvs(1000)
plt.hist(sample, normed=True)

#Тут строим теоретическую плотность вероятности
x = np.linspace(0,4,100)
pdf = distr.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

В распределении Рэлея матожидание и дисперсия считаются по следующим формулам (параметр распределения $sigma$ считаю равным 1):
$$EX = {\sqrt{\pi/2}}$$
$$DX = 2 -\pi/2$$

```python
#Выборки из 5 элементов
expect = []
for i in range(1000):
    nabor = np.random.choice(sample, 5, 0.001)
    sum = 0
    for item in nabor:
        sum = sum + item
    expect.append(float(sum)/len(nabor))
plt.hist(expect, normed=True)

#Матожидание и дисперсия, которыми можно приблизить выборочное среднее по ЦПТ
parameter_1 = math.sqrt(math.pi/2)
parameter_2 = math.sqrt((2 - math.pi/2)/5)

#Поверх рисуем плотность нормального распределения
norm_rv = sts.norm(parameter_1,parameter_2)
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
