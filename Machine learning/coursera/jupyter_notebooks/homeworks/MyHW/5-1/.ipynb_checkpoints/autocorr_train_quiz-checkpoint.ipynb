```python
import pandas as pd
import statsmodels.api as sm
%pylab inline
```

```python
milk = pd.read_csv('monthly-milk-production.csv', sep=';', index_col=['month'], parse_dates=['month'], dayfirst=True)
milk.head()
```

```python
milk.plot(figsize=(10, 8));
```
