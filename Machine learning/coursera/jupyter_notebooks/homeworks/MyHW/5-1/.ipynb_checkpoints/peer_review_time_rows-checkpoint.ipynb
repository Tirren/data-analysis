# Прогнозирование уровня средней заработной платы в России

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
import itertools
from scipy import stats
%pylab inline
```

```python
# Функция для обратного преобразования Бокса-Кокса
def invboxcox(y, lmbda):
      return np.exp(y) if lmbda == 0 else np.exp(np.log(lmbda*y + 1) / lmbda)
```

```python
row = pd.read_csv('WAG_C_M.csv', sep=';', index_col=['month'], parse_dates=['month'], dayfirst=True)
row.head()
```

Посмотрим визуально на ряд.

```python
row['WAG_C_M'].plot(figsize=(16, 9));
```

Видно, что у ряда есть ярко выраженный возрастающий тренд и сезонность. Также видно, что дисперсия нестабильна и со временем возрастает. Ряд нестационарен, проверим это формально и построим декомпозицию.

```python
plt.figure(figsize=(16, 10))
sm.tsa.seasonal_decompose(row['WAG_C_M']).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(row['WAG_C_M'])[1])
```

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии.

```python
row['WAG_C_M_box'], lmbda = stats.boxcox(row['WAG_C_M'])
row['WAG_C_M_box'].plot(figsize=(16, 9));
plt.ylabel(u'Transformed salary')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(row['WAG_C_M_box'])[1])
```

Избавимся от сезонности и тренда. Применим сезонное дифференцирование, после чего снова посмотрим SLT-декомпозицию и проверим стационарность.

```python
row['WAG_C_M_box_diff'] = row['WAG_C_M_box'] - row['WAG_C_M_box'].shift(12)

plt.figure(figsize=(16, 10))
sm.tsa.seasonal_decompose(row['WAG_C_M_box_diff'].dropna()).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(row['WAG_C_M_box_diff'].dropna())[1])
```

Критерий Дики-Фуллера отвергает гипотезу об стационарности ряда, однако визуально всё ещё виден убывающий тренд. Добавим обычное дифференцирование.

```python
row['WAG_C_M_box_diff2'] = row['WAG_C_M_box_diff'] - row['WAG_C_M_box_diff'].shift(1)

plt.figure(figsize=(16, 10))
sm.tsa.seasonal_decompose(row['WAG_C_M_box_diff2'].dropna()).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(row['WAG_C_M_box_diff2'].dropna())[1])
```

Критерий Дики-Фуллера по прежнему отвергает гипотезу нестационарности. Визуально ряд выглядит лучше. Для дальнейшего построения модели подберём начальные приближения для p, P, q, Q. Для этого посмотрим на ACF и PACF полученного ряда.

```python
plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
sm.graphics.tsa.plot_acf(row['WAG_C_M_box_diff2'].dropna().values.squeeze(), lags=48, ax=ax)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 2)
sm.graphics.tsa.plot_pacf(row['WAG_C_M_box_diff2'].dropna().values.squeeze(), lags=48, ax=ax)
plt.show()
```

Начальные приближения: $p = 1, P = 1, q = 5, Q = 0.$ Визуально кажется, что некоторые значения лежат на границе коридора. Из-за особенностей критерия Акаике я решил не включать слишком большие значения параметров, визуально лежащих на границе ($p = 10, q = 10$)

```python
# Однако, на всякий случай, расширим немного область перебираемых значений.
ps, Ps = range(0, 2), range(0, 2)
qs, Qs = range(0, 6), range(0, 2)
d, D = 1, 1
```

```python
parameters = itertools.product(ps, qs, Ps, Qs)
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for parameter in list(parameters):
    # Модель обучается не на всех наборах параметров
    try:
        model=sm.tsa.statespace.SARIMAX(row['WAG_C_M_box'], order=(parameter[0], d, parameter[1]), 
                                        seasonal_order=(parameter[2], D, parameter[3], 12)).fit(disp=-1)
    # Выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', parameter)
        continue
    aic = model.aic
    # Сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_parameter = parameter
    results.append([parameter, model.aic])
    
warnings.filterwarnings('default')
```

```python
# Посмотрим на первых пять лучших моделей
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
result_table.sort_values(by='aic', ascending=True).head()
```

Лучшая модель:

```python
print(best_model.summary())
```

Проанализируем остатки лучшей модели.

```python
plt.figure(figsize=(16, 9))
plt.subplot(2, 1, 1)
best_model.resid.dropna().plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(2, 1, 2)
sm.graphics.tsa.plot_acf(best_model.resid.dropna().values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid.dropna(), 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid.dropna())[1])
```

Остатки несмещены (подтверждается критерием Стьюдента), стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой). Посмотрим, насколько хорошо модель описывает данные.

```python
row['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(16, 9))
row['WAG_C_M'].plot()
row['model'].dropna().plot(color='r')
plt.ylabel('Mean salary')
plt.show()
```

Красная линия - модельные данные. Видно что данные описываются достаточно неплохо. Построим прогноз на два года вперёд.

```python
row['WAG_C_M'].plot(figsize=(16, 9))
invboxcox(best_model.forecast(24), lmbda).plot(color='r');
```

Красная линия - прогноз на следующие два года. График адекватный, можно сделать вывод, что модель выбрана и работает корректно.
